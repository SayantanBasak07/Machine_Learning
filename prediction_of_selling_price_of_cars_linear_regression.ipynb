```python
df.replace({"Fuel_Type":{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
df.replace({"Seller_Type":{'Dealer':0,'Individual':1}},inplace=True)
df.replace({"Transmission":{'Manual':0,'Automatic':1}},inplace=True)
```

```python
#splitting
x=df.drop(["Car_Name","Selling_Price"],axis=1)
y=df["Selling_Price"]
```

```python
from sklearn.model_selection import train_test_split  #library for splitting data for test and train
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics  #for accuracy and error detection
```

```python
#load the model
lr=LinearRegression()
```

```python
lr
```

```python
lr.fit(x_train,y_train)
```

```python
train_pred=lr.predict(x_train)
test_pred=lr.predict(x_test)
```

```python
test_pred
```

```python
#squared error
error_train=metrics.r2_score(y_train,train_pred)
error_train
```

```python
error_test=metrics.r2_score(y_test,test_pred)
error_test
```

```python
plt.scatter(train_pred,y_train)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
```
